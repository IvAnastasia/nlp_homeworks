###**Speech to Text:** Vosk (библиотеки vosk и wave)

Скачала [отсюда](https://https://alphacephei.com/vosk/models) модель **vosk-model-small-de-0.15** для немецкого и загрузила в гугл коллаб в виде архива.

```python
! unzip vosk-model-small-de-0.15.zip
```

```python
from vosk import Model, KaldiRecognizer
import wave
```

```python
wf = wave.open('arendt.wav', "rb")

results = ""
textResults = []

model = Model(r"/content/vosk-model-small-de-0.15")
recognizer = KaldiRecognizer(model, wf.getframerate())
recognizer.SetWords(True)
```

```python
while True:
    data = wf.readframes(4000)
    if len(data) == 0:
        break
    if recognizer.AcceptWaveform(data):
        recognizerResult = recognizer.Result()
        results = results + recognizerResult
        resultDict = json.loads(recognizerResult)
        textResults.append(resultDict.get("text", ""))

results = results + recognizer.FinalResult()
resultDict = json.loads(recognizer.FinalResult())
textResults.append(resultDict.get("text", ""))
```

```python
vosk_text = " ".join(textResults)
```

```python
vosk_text
```

```python
with open('vosk_text.txt', 'w') as file:
    file.write(vosk_text)
```

###**Оценка:** Word Error Rate (библиотека jiwer)

```python
from jiwer import wer
import pandas as pd
```

Беру whisper текст и вручную локально редактирую его, создаю стандарт разметки, опираясь на расшифроку [тут](https://https://www.rbb-online.de/zurperson/interview_archiv/arendt_hannah.html).

```python
with open('golden.txt') as file:
  golden = file.read()
```

```python
golden
```

```python
eval = pd.DataFrame(columns=['model', 'WER'])
```

```python
model_list = ['youtube', 'whisper', 'google', 'vosk']

for model in model_list:
  with open(model + '_text.txt') as file:
    text = file.read()
  er = wer(golden.lower(), text.lower())
  row = [model, er]
  eval.loc[len(eval),:] = row
```

```python
eval.sort_values(by=['WER'])
```

Посмотрим на качество в отсутствии знаков препинания (т. к. youtube и vosk модели их не расставляли)

```python
import string
```

```python
eval['WER_no_punct'] = wer_new
```
