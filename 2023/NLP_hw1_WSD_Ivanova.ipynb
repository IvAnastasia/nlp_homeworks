```python
vm = adagram.VectorModel.load('all.a010.p10.d300.w5.m100.nonorm.slim.joblib')
```

```python
corpus_df = pd.DataFrame(columns=["word", "context", "context_tokenized", "Ada_meaning"])
```

Из-за того, что изначальный корпус в формате conllu, полные тексты предложений необходимо собирать из слов. Код ниже записывает полный текст и лемматизированный текст (так как он уже есть в conllu) в датафрейм. Чтобы восстановить пробелы, я пользуюсь (несовершенным) правилом: пробелы ставятся между двумя словами и между знаком препрепинания и словом, если этот знак не открывающаяся кавычка.

```python
for word in corpus_words:
  for sent in corpus[word]:
    context = ''
    context_lemmatized = ''
    last = 0
    for token in sent.tokens:
      current = token.pos
      space = False
      if current != 'PUNCT' and last != 'PUNCT' and last != 0:
        space = True
      elif current != 'PUNCT' and last == 'PUNCT' and last != "«":
        space = True
      if space == True:
        context += " "
        context_lemmatized += " "
      context += token.text
      context_lemmatized += token.lemma
      last = token.pos
    means = vm.disambiguate(word, context_lemmatized)
    i = np.argmax(means)
    corpus_df = pd.concat([pd.DataFrame([[word, context, context_lemmatized, i]], columns=corpus_df.columns), corpus_df], ignore_index=True)
```

####ElMO

```python
morph = MorphAnalyzer()
token = RegexpTokenizer('\w+')
stops = set(stopwords.words('russian'))

def normalize(text):
    text = text.lower()
    words = [morph.parse(word)[0].normal_form for word in tokenize(text) if word]
    return words

def tokenize(text):
    return token.tokenize(text)
```

Функцию get_elmo_vectors из семинарской тетрадки немного модифицируем, чтобы в ней учитывались ошибки при токенизации (чтобы наличие ошибок не сбивало порядок контекстов, при этом не влияя на результат кластеризации)

```python
def get_elmo_vectors(word, contexts, model):
    tokens = [normalize(c) for c in contexts]
    all_vectors = model.get_elmo_vectors(tokens)
    word_vecs = []
    errors = []
    for i in range(len(contexts)):
        try:
            word_vecs.append(all_vectors[i][tokens[i].index(word)])
        except ValueError:
            errors.append(i)
            #список индексов ошибочных контекстов
    return word_vecs, errors
```

```python
model = ElmoModel()
model.load("196")
```

Как метод кластеризации будем использовать KMeans, количество кластеров зададим равным количеству значений слова по словарю.

```python
grouped_df = corpus_df.groupby('word')[['word', 'context']]

for key, _ in grouped_df:
    texts = grouped_df.get_group(key)['context']
    X, err = get_elmo_vectors(key, texts, model)

    cluster = KMeans(n_clusters= len(definitions.loc[definitions["lexeme"] == key]), random_state=0, n_init="auto")
    cluster.fit(X)
    labels = np.array(cluster.labels_) + 1

    indx0 = corpus_df.loc[corpus_df["word"] == key].index[0]
    indx1 = corpus_df.loc[corpus_df["word"] == key].index[-1]

    for i in err:
      labels = np.insert(labels, i, -1)
      #припишем -1 контекстам, для которых не получился эмбеддинг

    corpus_df.loc[indx0:indx1, 'elmo'] = labels.astype(int)
```

###ElMo

Для ElMo эмбеддингов из текста необходимо, чтобы текст содежал ключевое слово (значения которого мы разделяем), поэтому добавим к текстам в начале определяемое слово.

```python
grouped_df = definitions.groupby('lexeme')[['lexeme', 'meaning']]

for key, _ in grouped_df:
    texts = []
    for text in grouped_df.get_group(key)['meaning']:
      texts.append(key + " — " + text)
      #добавляем в текст определяемое слово
    X, err = get_elmo_vectors(key, texts, model)

    cluster = KMeans(n_clusters=len(definitions.loc[definitions["lexeme"] == key]), random_state=0, n_init="auto")
    cluster.fit(X)
    labels = np.array(cluster.labels_) + 1

    indx0 = definitions.loc[definitions["lexeme"] == key].index[0]
    indx1 = definitions.loc[definitions["lexeme"] == key].index[-1]

    for i in err:
      labels = np.insert(labels, i, -1)

    definitions.loc[indx0:indx1, 'elmo'] = labels.astype(int)
```

##Сравниваем методы WSD

По метрике **AdaGram справился значительно лучше, чем Elmo эмбеддинги**, но и у AdaGram accuracy не достигает 60%, хотя я размечала спорные случаи в пользу модели.

**Очевидная проблема:** сопоставление разметки контекстов со словарными определениями через автоматическое приписывание значений определениям.

Определение слова не является типичным контекстом для появления слова, в каком-то смысле -- наоборот крайне нетипичным (потому что определение и слово взаимозаменяемы до определенной степени), поэтому у Elmo эмбеддингов, например, такая низкая accuracy. При ручной разметке я заметила, что часто слова правилльно объединялись в кластеры, но определение с ними не соотносилось -- возможно, потому что такой способ приписывания значений не самый удачный.
Кроме того, словарные значения иногда не кажутся соответствующими современным (например, у слова *правый* есть значение "Враждебный политике партии внутри самой партии" [КПСС]).


В целом я бы не считала accuracy показательной на такой маленькой выборке (даже если мы изменим принцип сопоставления разметки с определениями).
