**pymorphy**

```python
! pip install pymorphy2
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

–¢–∞–∫ –∫–∞–∫ —É pymorphy –Ω–µ—Ç —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä–∞, –≤–æ–∑—å–º—ë–º —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä –∏–∑ nltk:

```python
pymorphy_tokens = []
pymorphy_tags = []
for text in texts:
    tokens = nltk.word_tokenize(text)
    for token in tokens:
        pymorphy_tokens.append(token)
        parsed = morph.parse(token)[0]
        pymorphy_tags.append(parsed.tag.POS)
```

```python
pymorphy_df = pd.DataFrame(columns = ['token', 'tag'])
pymorphy_df['token'] = pymorphy_tokens
pymorphy_df['tag'] = pymorphy_tags
```

**Natasha**

```python
! pip install natasha
```

```python
from natasha import (
    Segmenter,
    MorphVocab,
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    Doc
)
```

```python
segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
```

```python
natasha_tokens = []
natasha_tags = []

tokens = []
for text in texts:
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    for token in doc.tokens:
        natasha_tokens.append(token.text)
        natasha_tags.append(token.pos)
```

```python
! pip install stanza
```

```python
import stanza
stanza.download('ru')
```

```python
nlp = stanza.Pipeline(lang='ru', processors='tokenize,pos')
```

```python
stanza_tokens = []
stanza_tags = []

for text in texts:
    doc = nlp(text)
    for sent in doc.sentences:
        for word in sent.words:
            stanza_tokens.append(word.text)
            stanza_tags.append(word.upos)
```

```python
def accuracy(model, df):
    right_counter = 0
    total = 0
    for i in range(len(df)):
        try:
            if df.loc[i]['tag'] != 'X' and df.loc[i]['tag'] != 'NUM' and df.loc[i]['tag'] != 'PUNCT':
                if df.loc[i]['tag'] == df.loc[i][model]:
                    right_counter += 1
                if model == 'pymorphy': 
                    if (df.loc[i]['tag'] == 'SCONJ' or df.loc[i]['tag'] == 'CCONJ') and (df.loc[i][model] == 'CONJ'):
                        right_counter += 1
                    if df.loc[i]['tag'] == 'PROP' and df.loc[i][model] == 'NOUN':
                        right_counter += 1
                if df.loc[i][model] != None and df.loc[i][model] != 'NaN' and df.loc[i][model] != 'PUNCT' and df.loc[i][model] != 'NUM':
                    total += 1
        except ValueError:
            pass
        #pymorphy —Ä—É–≥–∞–µ—Ç—Å—è, —á—Ç–æ –≤ –µ–≥–æ –∫–æ–ª–æ–Ω–∫–µ –µ—Å—Ç—å —Ç–µ–≥–∏ –Ω–µ –∏–∑ –µ–≥–æ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ —Å–µ—Ç–∞ –∏ NaN-—ã, –ø–æ—ç—Ç–æ–º—É –Ω—É–∂–Ω–æ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError. –ö–∞—á–µ—Å—Ç–≤–æ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π –æ—Ç–ª—á–∏–Ω–æ —Å—á–∏—Ç–∞–µ—Ç—Å—è –∏ –±–µ–∑ —ç—Ç–æ–≥–æ.
    return right_counter/total
```

```python
def chunker(seq, size):
    output = []
    for pos in range(0, len(seq), size):
        output.append(seq[pos:pos + size])
    return output
```

```python
def chunker(seq, size):
    output = []
    for element in range(0, len(seq)-size+1):
        output.append(seq[element:element + size])
    return output
```

```python
chunker([1, 2, 3, 4, 5], 2)
```

```python
def ngrams(text):
    res = []
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    for group in chunker(doc.tokens, 2):
        if group[0].pos == 'ADV' and group[1].pos == 'ADJ':
            res.append(' '.join([group[0].text, group[1].text]))
        if group[0].pos == 'ADJ' and group[1].pos == 'NOUN':
            res.append(' '.join([group[0].text, group[1].text]))
    for group in chunker(doc.tokens, 3):
        if group[0].pos == 'VERB' and group[1].pos == 'ADP' and group[2].pos == 'NOUN':
            res.append(' '.join([group[0].text, group[1].text, group[2].text]))
    return res
```
